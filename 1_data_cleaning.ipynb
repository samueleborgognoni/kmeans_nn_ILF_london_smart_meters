# Data cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')

drive_folder_path = '/content/drive/MyDrive/-CareerðŸ“š/_UNIVERSITY/__Macchine_ed_azionamenti_elettrici/_Progetto_MAE/mae_proj/'
```

```python
# Import first block of data
df = pd.read_csv(drive_folder_path + 'data/block_0.csv') # init
# Add to the df new blocks of data from different files
for i in range(1,15):
  # df.iloc[len(df)] = pd.read_csv(drive_folder_path + f'data/block_{i}.csv')
  df_i = pd.read_csv(drive_folder_path + f'data/block_{i}.csv')
  df = pd.concat([df, df_i], ignore_index=True)

df.shape
```

```python
# Display the header of DF
df.head(5)
```

```python
# Count the number of single LCLid (housholds)
df['LCLid'].nunique()
```

```python
# Count the number of samples for a given ID
id_count = df['LCLid'].value_counts() # type: pd.series
print(f"{id_count[:5]}\n{'-'*35}")

# Evaluate the variance of the number of samples for each ID
print(f"ID_count MEAN: {int(id_count.mean())}")
print(f"ID_count STD: {int(id_count.std())}")
print(f"ID_count MAX: {max(id_count)} \t -- ID: {id_count.argmax()}")
print(f"ID_count MIN: {min(id_count)} \t -- ID: {id_count.argmin()}")
```

```python
df.head()
```

```python
# Count the NaN values
print(f"NaN values in 'energy(kWh/hh)': {len(df[df['energy(kWh/hh)'].isna()])}")
print(f"NaN values in 'tstp': {len(df[df['tstp'].isna()])}")
```

```python
# Convert 'tstp' to datetime objects
df['tstp'] = pd.to_datetime(df['tstp'])

# Convert 'energy(kWh/hh)' to numeric, replacing non numeric values to NaN
df['energy(kWh/hh)'] = pd.to_numeric(df['energy(kWh/hh)'], errors='coerce')
```

```python
# Save the cleaned dataset into a new .csv file
df.to_csv(drive_folder_path + 'data/df_cleaned.csv', index=False)
```
